```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Задание 17
rooms = set(df_refr.Rooms.astype('int'))
data = [len(df_refr[df_refr.Rooms == i].Rooms) for i in rooms]
maxval = max(data)
explode = []
for i in range(len(rooms)):
    if data[i] == maxval:
        explode.append(0.1)
    else:
        explode.append(0.0)


fig, ax = plt.subplots(figsize=(14,7))
ax = plt.pie(x=data, labels=rooms, autopct='%1.1f%%', explode=explode, labeldistance=None, colors = ['aquamarine', 'turquoise', 'lightseagreen', 'mediumaquamarine', 'paleturquoise', 'cyan', 'cadetblue', 'darkturquoise'])
plt.title('Процентное соотношение кол-ва квартир в зависимости от количества комнат', fontsize=18)
plt.legend(title='Кол-во комнат', loc='upper right');
```

```python
# Задание 19
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,8))

sns.scatterplot(data=df_refr, x='Price', y='Rooms', ax=ax[0, 0])
sns.scatterplot(data=df_refr, x='Price', y='Square', ax=ax[1, 0])
sns.scatterplot(data=df_refr, x='Price', y='HouseFloor', ax=ax[0, 1])
sns.scatterplot(data=df_refr, x='Price', y='HouseYear', ax=ax[1, 1])

plt.suptitle('Диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price ');
```

```python
# Задание 22
columns = ['Rooms', 'Square', 'HouseFloor', 'HouseYear']

g = sns.PairGrid(data=df_refr[df_refr['HouseYear']<=datetime.datetime.now().year], x_vars=columns, y_vars=columns, diag_sharey=False)

g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_upper(sns.scatterplot)

plt.show()
```
